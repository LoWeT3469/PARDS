```python
# CDGR / AVEAA frequency domain demo (code from https://realpython.com/python-scipy-fft/)

from scipy.fft import rfft, rfftfreq
from scipy.signal import find_peaks
import numpy as np
import matplotlib.pyplot as plt

# Perform Fourier transform
x = subset_df["CDGR - iPress Wave"].values  # Input signal
# x = subset_df["AVEA - PIP"].values  # Input signal
fs = 200  # Sampling frequency (CDGR pressure)
# fs = 0.2  # Sampling frequency

# Number of samples in x
n = len(x)

# Compute the Fourier transform
yf = rfft(x)   # The amplitude
yf = 20 * np.log10(yf)   # Convert amplitude to gain in decibels (dB)
xf = rfftfreq(n, 1 / fs)

# Define the prominence values to test
prominences = [5, 10, 15, 20, 25]

for prominence in prominences:
    # Find peaks in the frequency spectrum
    peaks, _ = find_peaks(np.abs(yf), prominence=prominence)
    
    # Calculate the derivative of the gain curve
    dyf = np.gradient(yf, xf)

    # Find the peak corresponding to the "elbow" area
    elbow_peak_idx = np.argmax(dyf[peaks])

    # Find the minimum peak
    min_peak_idx = np.argmin(np.abs(yf[peaks]))

    # Plot the frequency content
    plt.figure(figsize=(10, 5))
    plt.plot(xf, np.abs(yf))
    plt.xlabel('Frequency [Hz]')
    plt.ylabel('Gain (dB)')
    plt.title(f'Frequency Content (Prominence={prominence})')
    plt.grid(True)
    
    # Annotate peaks in the plot
    for peak in peaks:
        plt.plot(xf[peak], np.abs(yf[peak]), 'ro', markersize=2)  # Plot a red dot at the peak position

    # Annotate maximum peak
    max_peak_idx = np.argmax(np.abs(yf[peaks]))
    plt.annotate(f'Max Peak: {xf[peaks[max_peak_idx]]:.2f} Hz', 
                 xy=(xf[peaks[max_peak_idx]], np.abs(yf[peaks[max_peak_idx]])), 
                 xytext=(10, 20),
                 textcoords='offset points',
                 arrowprops=dict(arrowstyle="->", connectionstyle="arc3"),
                 color='red',
                 fontsize=16)

    # Annotate minimum peak at the "elbow" area
    plt.annotate(f'Elbow Peak: {xf[peaks[elbow_peak_idx]]:.2f} Hz', 
                 xy=(xf[peaks[elbow_peak_idx]], np.abs(yf[peaks[elbow_peak_idx]])), 
                 xytext=(10, 60),
                 textcoords='offset points',
                 arrowprops=dict(arrowstyle="->", connectionstyle="arc3"),
                 color='blue',
                 fontsize=16)

    # Annotate minimum peak
    plt.annotate(f'Min Peak: {xf[peaks[min_peak_idx]]:.2f} Hz', 
                 xy=(xf[peaks[min_peak_idx]], np.abs(yf[peaks[min_peak_idx]])), 
                 xytext=(10, 100),
                 textcoords='offset points',
                 arrowprops=dict(arrowstyle="->", connectionstyle="arc3"),
                 color='red',
                 fontsize=16)

    plt.show()
```
