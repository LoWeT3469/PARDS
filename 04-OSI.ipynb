### 1. "58" cohort, CDGR ventilator, 39 patients, 1st hour OSI.

```python
import pandas as pd
import os

# Define the directory containing the CSV files
directory = '/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/02-EVENTS/a-CDGR+AVEAA+SVU(58)_20240205/a-58_CDGR_39_2nd/'

# Initialize empty lists to store start times, OSI, and file names
OSIs = []

# Iterate over each CSV file in the directory
for i in range(1, 40):
    filename = f'a_58_CDGR_39_2nd ({i}).csv'
    file_path = os.path.join(directory, filename)
    
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    
    # Convert 'Time' column to datetime format, handle errors if any
    try:
        df['Time'] = pd.to_datetime(df['Time'], format='%Y-%m-%d %H:%M:%S.%f')
    except Exception as e:
        print(f"Error converting 'Time' column to datetime format in file: {filename}")
        print(e)
        start_times.append({'Start Time': None, 'File Name': filename})
        continue

    # Check if the DataFrame has non-null values for the required columns
    if df[['Oxygenation Saturation Index']].notna().all(axis=1).any():
        # Find the index where the condition is met for the first time
        OSI_start_index = df[['Oxygenation Saturation Index']].notna().all(axis=1).idxmax()
        # Find the starting time for the OSI
        OSI_start_time = df.loc[OSI_start_index, 'Time']
        # Find the specific value for the OSI
        OSI_value = df.loc[OSI_start_index, 'Oxygenation Saturation Index']
        # Append start time, OSI, and file name to the list
        OSIs.append({'Start Time': OSI_start_time, 'OSI': OSI_value, 'File Name': filename})

    else:
        OSIs.append({'Start Time': None, 'OSI': None, 'File Name': filename})

# Convert the lists to DataFrames
OSIs_df = pd.DataFrame(OSIs)

# Save the DataFrames to CSV files
OSIs_df.to_csv('OSI_58_CDGR_39_1st.csv', index=False)
```
