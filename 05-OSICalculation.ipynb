### OSI Calculation:

**Oxygenation Saturation Index (OSI)** $= \frac{Paw \space \times \space FiO_{2}}{SpO_{2}}$

```python
# Concate 'a_58_CDGR_39_1st (35)_p1.csv' and 'a_58_CDGR_39_1st (35)_p2.csv'

import pandas as pd

# Paths to the CSV files
file1_path = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/02-EVENTS/a-CDGR+AVEAA+SVU(58)_20240205/a-58_CDGR_39_1st/a-58_CDGR_39_1st_Raw/a_58_CDGR_39_1st (35)_p1.csv"
file2_path = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/02-EVENTS/a-CDGR+AVEAA+SVU(58)_20240205/a-58_CDGR_39_1st/a-58_CDGR_39_1st_Raw/a_58_CDGR_39_1st (35)_p2.csv"

# Read CSV files into pandas DataFrames
df1 = pd.read_csv(file1_path)
df2 = pd.read_csv(file2_path)

# Concatenate the DataFrames with p1.csv appearing before p2.csv
concatenated_df = pd.concat([df1, df2], ignore_index=True)

# Save the concatenated DataFrame to a new CSV file
output_file_path = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/02-EVENTS/a-CDGR+AVEAA+SVU(58)_20240205/a-58_CDGR_39_1st/a-58_CDGR_39_1st_Raw/a_58_CDGR_39_1st (35).csv"
concatenated_df.to_csv(output_file_path, index=False)
```
