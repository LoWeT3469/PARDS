#### Load 58 patients datasets

- ***58*** patients data from Oct. 2023 to Dec. 2023, should have all seven features (pressure, flow, volumn, OSI, SpO2, FiO2, Paw)
- 12hrs(before starting date) + 24 hours + 12hrs(after starting date) data for each patient's starting date, need to decide the starting point, which is all seven features showed up.
- one of three ventilators (CDGR or AVEAA or SVU).

```python
# Read 58 csv files iteratively, and summary the statistics for each file.
import pandas as pd

# Initialize a pandas ExcelWriter
with pd.ExcelWriter('a_58_ext_column_statistics.xlsx') as writer:
    # Iterate over the range of file numbers
    for i in range(1, 233):
        # Generate the filename
        filename = f'/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/02-EVENTS/a-CDGR+AVEAA+SVU(58)_20240205/a-58_Extension/a_58_ext_{i:03}.csv'
        # Read the CSV file into a DataFrame
        df = pd.read_csv(filename)
        # Get the number of rows in the DataFrame
        num_rows = len(df)
        # Initialize a dictionary to store column-wise statistics
        column_stats = {}
        # Iterate over each column in the DataFrame
        for col in df.columns:
            # Get the number of non-empty values in the column
            non_empty_values = df[col].count()
            # Get the number of empty values in the column
            empty_values = num_rows - non_empty_values
            # Store the column statistics in the dictionary
            column_stats[col] = {'num_rows': num_rows, 'non_empty_values': non_empty_values, 'empty_values': empty_values}
        # Convert the column statistics to a DataFrame
        stats_df = pd.DataFrame(column_stats)
        # Add the DataFrame to the ExcelWriter with the file name as the sheet name
        stats_df.to_excel(writer, sheet_name=f'sheet_{i:03}', index=False)
```

```python
# Concatenate all 58 files' statistics to one table
import pandas as pd

# Initialize an empty list to store the DataFrames
dfs = []

# Iterate over the range of sheet numbers
for i in range(1, 233):
    # Generate the sheet nameZ
    sheet_name = f'sheet_{i:03}'
    # Read the sheet into a DataFrame
    df = pd.read_excel('a_58_ext_column_statistics.xlsx', sheet_name=sheet_name)
    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate the DataFrames vertically by stacking them by rows
result_df = pd.concat(dfs, axis=0)

# Generate the row names for every third row
row_names = []
for i in range(1, 233):
    row_names.extend([f'num_rows_{i:03}', f'non_empty_values_{i:03}', f'empty_values_{i:03}'])
    
# Insert the row names into the DataFrame
result_df.index = row_names + [f'extra_{i}' for i in range(len(result_df) - len(row_names))]

# Save the concatenated DataFrame to a CSV file
result_df.to_csv('a_58_ext_column_statistics_concatenated_with_row_names.csv')
```

```python
# Add "Group" column:
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("a_58_ext_column_statistics_concatenated_with_row_names.csv")

# Rename the first column as "File Name"
df.rename(columns={'Unnamed: 0': 'File Name'}, inplace=True)

# Keep only the rows where the "File Name" starts with "non"
df = df[df['File Name'].str.startswith('non')]

# Add a new column "Group" with group numbers
group_size = 4
df['Group'] = (df.index // group_size) + 1

# Save the DataFrames to CSV files
df.to_csv('a_58_ext_column_statistics_concatenated_with_row_names.csv', index=False)
```

```python
# Concatenate all 302 files' statistics to one table
import pandas as pd

# Initialize an empty list to store the DataFrames
dfs = []

# Iterate over the range of sheet numbers
for i in range(1, 605):
    # Generate the sheet nameZ
    sheet_name = f'sheet_{i:03}'
    # Read the sheet into a DataFrame
    df = pd.read_excel('b_151_ext_column_statistics.xlsx', sheet_name=sheet_name)
    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate the DataFrames vertically by stacking them by rows
result_df = pd.concat(dfs, axis=0)

# Generate the row names for every third row
row_names = []
for i in range(1, 605):
    row_names.extend([f'num_rows_{i:03}', f'non_empty_values_{i:03}', f'empty_values_{i:03}'])
    
# Insert the row names into the DataFrame
result_df.index = row_names + [f'extra_{i}' for i in range(len(result_df) - len(row_names))]

# Save the concatenated DataFrame to a CSV file
result_df.to_csv('b_151_ext_column_statistics_concatenated_with_row_names.csv')
```

```python
# Add "Group" column:
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("b_151_ext_column_statistics_concatenated_with_row_names.csv")

# Rename the first column as "File Name"
df.rename(columns={'Unnamed: 0': 'File Name'}, inplace=True)

# Keep only the rows where the "File Name" starts with "non"
df = df[df['File Name'].str.startswith('non')]

# Reset the index to re-number from 0 to length(df)
df.reset_index(drop=True, inplace=True)

# Add a new column "Group" with group numbers
group_size = 4
df['Group'] = (df.index // group_size) + 1

# Save the DataFrames to CSV files
df.to_csv('b_151_ext_column_statistics_concatenated_with_row_names.csv', index=False)
```
