```python
# CDGR (indiviual sample)

import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Directory path
## 24hrs version:
directory_path = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/02-EVENTS/PvsV_24hrs_Raw/CDGR/"
## Raw version:
# directory_path = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/02-EVENTS/PvsV_1hr_Raw/CDGR/"
# ## Cleaned version:
# directory_path = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/02-EVENTS/PvsV_1hr_Cleaned/CDGR/"

# filename = "a_58_CDGR_39_24hrs (32).csv"
filename = "b_151_CDGR_11_24hrs (11).csv"

filepath = os.path.join(directory_path, filename)

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(filepath)

df['Time'] = pd.to_datetime(df['Time'], format='%Y-%m-%d %H:%M:%S.%f')

# Rename columns
df.rename(columns={
    'CDGR - Set: Pinsp': 'PIP_Settings',
    'CDGR - Set: PEEP L': 'PEEP_Settings',
    'CDGR - PIP': 'PIP_Actual',
    'CDGR - PEEP': 'PEEP_Actual',
    'CDGR - eVT': 'Tidal_Volume'
}, inplace=True)

## Add PID to the title
# Step 1: Read the CSV file into a DataFrame
# df1 = pd.read_csv("Mode_Cat_CDGR_58.csv")
df1 = pd.read_csv("Mode_Cat_CDGR_151.csv")

# Step 2: Check if "filename" matches "File Name 2" and add PID to title
matched_row1 = df1[df1["File Name 2"] == filename]

if not matched_row1.empty:
    pid = matched_row1["PID"].iloc[0]  # Extract PID from filename
    print(pid)
    title = f'CDGR Ventilator (PID: {pid})'  # Create the title with PID
else:
    title = "CDGR Ventilator"
    
## Add patient weight according to the PID
# Step 1: Read the CSV file into a DataFrame
# df2 = pd.read_csv("58_CDGR_39_24hrs_Weights.csv")
df2 = pd.read_csv("151_CDGR_11_24hrs_Weights.csv")

# Step 2: Check if "filename" matches "File Name 2" and add PID to title
matched_row2 = df2[df2["patient_id"] == pid]

if not matched_row2.empty:
    weight = matched_row2["average_value"].iloc[0]  # Extract PID from filename
    print(weight)
else:
    weight = None

df['TV(mL/kg)'] = df['Tidal_Volume'] / weight
print(df['TV(mL/kg)'])

# Calculate average tidal volume at 8 am
eight_am_data = df[(df['Time'].dt.time >= pd.to_datetime('08:00:00').time()) & 
                   (df['Time'].dt.time <= pd.to_datetime('08:01:00').time())]

# Calculate average tidal volume at 8 am
eight_am_avg = eight_am_data['TV(mL/kg)'].mean()

# Create subplots with shared y-axis
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 8), sharey=True)

# Add a title for the entire figure including PID
fig.suptitle(title, fontsize=16)

# Violin plot
sns.violinplot(data=df, x='PIP_Settings', y='TV(mL/kg)', hue='PIP_Settings', palette='muted', ax=ax1, legend=False)
ax1.set_xlabel('PIP Settings (mbar)')
ax1.set_ylabel('Tidal Volume (mL/kg)')
ax1.set_title('Tidal Volume Distribution across PIP Settings')

# Add thresholds for Volume in violin plot
ax1.axhline(y=6, color='r', linestyle='--', linewidth=1, label='Tidal Volume = 6 mL/kg')
ax1.axhline(y=8, color='b', linestyle='--', linewidth=1, label='Tidal Volume = 8 mL/kg')

# Get unique PIP_Settings values for the violin plot, remove NaN, and sort from smallest to largest
unique_pip_settings = df['PIP_Settings'].dropna().unique()
unique_pip_settings.sort()
# Get the first row's "PIP_Settings" value from eight_am_data
first_row_pip_settings = eight_am_data.iloc[0]['PIP_Settings']
# Get the index of the first_row_pip_settings value in the unique_pip_settings list
index_of_first_row_pip_settings = list(unique_pip_settings).index(first_row_pip_settings)
# Plot average tidal volume at 8 am as red dot
ax1.scatter([index_of_first_row_pip_settings], [eight_am_avg], color='red', s=50, label=f'Average Tidal Volume at 8 am', zorder=10)
ax1.annotate(f'{eight_am_avg:.2f} mL/kg', xy=(index_of_first_row_pip_settings, eight_am_avg), xytext=(index_of_first_row_pip_settings+0.2, eight_am_avg+1),
             arrowprops=dict(shrink=0.1, width=1.5, headwidth=8, color='red'), color='red', fontsize=12)
ax1.legend()

# Box plot for volume at 8 am
sns.boxplot(data=eight_am_data, x='PIP_Settings', y='TV(mL/kg)', color='cyan', width=0.3, fliersize=0, ax=ax2)
ax2.set_xlabel('PIP Settings (mbar)')
ax2.set_ylabel('Tidal Volume (mL/kg)')
ax2.set_title('Tidal Volume at 8 am (1 minute)')

# Add thresholds for Volume in box plot
ax2.axhline(y=6, color='r', linestyle='--', linewidth=1, label='Tidal Volume = 6 mL/kg')
ax2.axhline(y=8, color='b', linestyle='--', linewidth=1, label='Tidal Volume = 8 mL/kg')
ax2.legend()

plt.show()
```

```python
# CDGR (all)

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Rename columns
result_df.rename(columns={
    'CDGR - Set: Pinsp': 'PIP_Settings',
    'CDGR - Set: PEEP L': 'PEEP_Settings',
    'CDGR - PIP': 'PIP_Actual',
    'CDGR - PEEP': 'PEEP_Actual',
    'CDGR - eVT': 'Tidal_Volume'
}, inplace=True)

# Print the count of NaN values in 'PIP_Settings' and 'Tidal_Volume'
print(result_df['PIP_Settings'].isna().sum())
print(result_df['Tidal_Volume'].isna().sum())

# Drop rows where 'PIP_Settings' is NaN
result_df.dropna(subset=['PIP_Settings'], inplace=True)

# Drop rows where 'Tidal_Volume' is NaN
result_df.dropna(subset=['Tidal_Volume'], inplace=True)

# Print the count of NaN values in 'PIP_Settings' and 'Tidal_Volume'
print(result_df['PIP_Settings'].isna().sum())
print(result_df['Tidal_Volume'].isna().sum())

# Iterate through each unique filename
for filename in result_df['Source_File'].unique():
    # Step 1: Read the CSV file into a DataFrame
    df1 = Mode_Cat_CDGR
    # Check if "filename" matches "File Name 2" and add PID to title
    matched_row1 = df1[df1["File Name 2"] == str(filename)]
    if not matched_row1.empty:
        pid = matched_row1["PID"].iloc[0]  # Extract PID from filename
        # print("PID:", pid)
        
        # Step 2: Read the weights CSV file into a DataFrame
        df2 = Weights_CDGR
        # Check if "filename" matches "patient_id" and add weight
        matched_row2 = df2[df2["patient_id"] == pid]
        if not matched_row2.empty:
            weight = matched_row2["average_value"].iloc[0]  # Extract weight from dataframe
            # print("Weight:", weight)
            
            # Calculate 'TV(mL/kg)' only if weight is not None
            if weight is not None:
                # Filter rows by filename
                subset_df = result_df[result_df['Source_File'] == filename].copy()  # Make a copy to avoid SettingWithCopyWarning

                # Calculate 'TV(mL/kg)' and assign to a new column using .loc
                subset_df.loc[:, 'TV(mL/kg)'] = subset_df['Tidal_Volume'] / weight

                # Filter rows where 'TV(mL/kg)' is less than 25
                subset_df = subset_df[subset_df['TV(mL/kg)'] < 25]
                # print(len(subset_df['TV(mL/kg)']))
                # print(subset_df['TV(mL/kg)'].isna().sum())
                
                # Update result_df with the calculated values using .loc
                result_df.loc[subset_df.index, 'TV(mL/kg)'] = subset_df['TV(mL/kg)']
                
                # Apply a 2-minute moving average to tidal volume
                df = subset_df.copy()  # Make a copy to avoid SettingWithCopyWarning
                df.set_index('Time', inplace=True)  # Set 'Time' as the index
                df.sort_index(inplace=True)  # Ensure the index is sorted to be monotonic
                df['TV(mL/kg)_ma'] = df['TV(mL/kg)'].rolling('2T').mean()  # Calculate the rolling mean with a 2-minute window
                df.reset_index(drop=True, inplace=True)  # Reset the index to original structure
                # print(len(df['TV(mL/kg)_ma']))
                # print(df['TV(mL/kg)_ma'].isna().sum())
                
                # Update result_df with the calculated values using .loc
                result_df.loc[subset_df.index, 'TV(mL/kg)_ma'] = df['TV(mL/kg)_ma'].values

                # Add PID and weight to the DataFrame using .loc
                result_df.loc[subset_df.index, 'PID'] = pid
                result_df.loc[subset_df.index, 'Weight'] = weight
            else:
                print("Weight is None, cannot calculate 'TV(mL/kg)'.")
        else:
            print("No weight information found for PID:", pid)
    else:
        print("No match found for filename:", filename)
```
